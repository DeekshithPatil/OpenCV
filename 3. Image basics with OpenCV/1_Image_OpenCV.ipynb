```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
img = cv2.imread('00-puppy.jpg') #this command directly reads the image as a numpy array
type(img)
```

```python
img.shape
```

```python
plt.imshow(img) 
```

```python
#The image is not displayed properly because, matplotlib and opencv indices for RGB are differen

#Matplotlib -> Red Green Blue
#OpenCV -> Blue Green Red
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #This command is used to convert the BGR image to RGB
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray) #still does not display grayscale image as matplotlib and cv2 used different indices for RGB
```

```python
plt.imshow(img_gray,cmap='gray') #this command is used to display the grayscale image 
```

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (1000,400)) #this command is used to resize the image of width 1000 and length 400
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img, w_ratio, h_ratio) #This command is used to scale the image height and width by -50%
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 0) #This command is used to flip the image about the horizontal axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 1) #This command is used to flip the image about the vertical axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, -1) #This command is used to flip the image about the horizontal axis and veritcal axis
plt.imshow(new_img)
```

```python
cv2.imwrite('totally_new.jpg',fix_img) #this command is used to save an image and returns true if successful
```

```python
fig = plt.figure(figsize=(12,12)) #This command is used to vary the display size of the image being displayed in the notebook
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```
