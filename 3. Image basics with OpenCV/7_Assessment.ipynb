```python
####Task: Read the dog_backpack.jpg image and display it in the correct RGB order using pyplot

import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline

img = cv2.imread('dog_backpack.jpg')

fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#Task: flip the image upside down
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
#Task: Draw an empty red triangle around the dog's face

new_img = cv2.rectangle(fix_img, (180,350), (650,750), (255,0,0),5)
plt.imshow(new_img)
```

```python
#Task: Draw an empty blue triangle centred around the dog's face

vertices = np.array([[450,350],[350,700],[600,700]])
pts = vertices.reshape(-1,1,2)
new_img = cv2.polylines(fix_img,[pts],False,(0,0,255),5)
plt.imshow(new_img)
```

```python
#Task: Fill out the triangle
new_img = cv2.fillPoly(fix_img, [pts], (0,0,255))
plt.imshow(new_img)
```
