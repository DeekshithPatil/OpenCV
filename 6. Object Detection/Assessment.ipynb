```python
#Task: Import the usual libraries that you think you'll need for car license plate recognition
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Task: Read in the car_plate.jpg image
img = cv2.imread('car_plate.jpg')
```

```python
#Task: Create a function that displays the image in a larger scale and correct coloring for matplotlib

def display(img):
    
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
display(img)
```

```python
#Task: Load the haarcascade_russian_plate_number.xml file

plate_detector = cv2.CascadeClassifier('haarcascades/haarcascade_russian_plate_number.xml')
```

```python
#Task: create a function that takes in an image and draws a rectangle around what it detects to be a license plate

def detect_plate(img):
    img_copy = img.copy()
    
    number_plate_detection = plate_detector.detectMultiScale(img,scaleFactor=1.2,minNeighbors = 5)
    
    for (x,y,w,h) in number_plate_detection:
        img_copy = cv2.rectangle(img_copy,(x,y),(x+w,y+h),(255,255,255),5)
        
    return img_copy
    
    
```

```python
result = detect_plate(img)
display(result)
```

```python
#Task: Blur the number plate

def blur_plate(img):
    img_copy = img.copy()
    
    number_plate_detection = plate_detector.detectMultiScale(img,scaleFactor=1.2,minNeighbors = 5)
    
    for x,y,w,h in number_plate_detection:
        #roi = img_copy[y:y+h,x:x+w]
        
        #roi = cv2.blur(roi,ksize=(21,21))
        
        #img_copy[y:y+h,x:x+w] = roi
        img_copy[y:y+h,x:x+w] = cv2.blur(img_copy[y:y+h,x:x+w],ksize=(15,15))
        
    return img_copy
```

```python
result = blur_plate(img)
display(result)
```
