```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

real_chess = cv2.imread('real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_RGB2GRAY)
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,65,0.01,10) #These command is used to perform Shi-Tomasi Corner detection
#The first parameter indicates the image on which the operation is to be performed
#The second parameter indicates the number of corners to be detected (no. of best corners) based on the quality level paramater at position 3
#Fourth parameter indicates the minimum distance of the corner
#put -1 (in 2nd parameter) if you want to detect all corners
#This function does not automatically mark the corners.
```

```python
corners = np.int0(corners) #Convert to integers
```

```python
for i in corners:
    x,y = i.ravel() #Flatten out the array
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```
