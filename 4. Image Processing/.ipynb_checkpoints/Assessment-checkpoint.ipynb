```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
#Task: Open and display Giraffes image
img = cv2.imread('giraffes.jpg')
show_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display(show_img)
```

```python
#Task: apply a binary threshold onto the image
img = cv2.imread('giraffes.jpg',0)
thresh,thresh_img =cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
display(thresh_img,cmap='gray')
```

```python
#Task: Open the giraffes img and convert its colorspace to HSV and display the image
img = cv2.imread('giraffes.jpg')
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
display(hsv)
```

```python
#Task: Create a low pass filter of size 4 x 4 with each value as 1/10 (type=float32) and then use the 2D convolution to blur the giraffes image
```

```python
img = cv2.imread('giraffes.jpg')
```

```python
kernel = np.ones((4,4),np.float32)/10
kernel
```

```python
blurred = cv2.filter2D(img,-1,kernel)
blurred = cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB)
```

```python
display(blurred)
```

```python
#Task: create a Sobel Filter along x, with a kernel size of 5 to the grayscale version of giraffes and then display it
```

```python
img = cv2.imread('giraffes.jpg',0)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display(sobelx,cmap='gray')
```

```python
#TASK: Plot the color histograms for the RED, BLUE, GREEN channel of the giraffe image. 
```

```python
img = cv2.imread('giraffes.jpg')
```

```python
color = ('b','g','r')
```

```python
for i,col in enumerate(color):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
```
