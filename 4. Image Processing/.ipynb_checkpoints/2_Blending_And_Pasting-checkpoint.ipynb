```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('dog_backpack.png')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('watermark_no_copy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
#BLENDING IMAGES OF SAME SIZE
```

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(src1=(img1),alpha=0.5,src2=(img2),beta=0.5,gamma=0) #This command is used to blend two images of the same size.
#Alpha and Beta add weights to image1 and image2
#It works only for images of same size
```

```python
plt.imshow(blended)
```

```python
#OVERLAY A SMALL IMAGE ON TOP OF A LARGER IMAGE (No Blending)
#Numpy Reassignment operation
```

```python
img1 = cv2.imread('dog_backpack.png')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('watermark_no_copy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0 #This values indicate where the starting position of the small image on the larger image should be
y_offset =0 

x_end = x_offset + small_img.shape[1] #These values are used to store the ending position of the smaller image on the larger image
y_end = y_offset + small_img.shape[0] 

#Note: Shape returns values in the format(y,x).
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
plt.imshow(large_img)
```
