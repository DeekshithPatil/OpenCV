```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('pothole_1.jpg',0)
display(img)
```

```python
img = cv2.blur(img,(3,3))
display(img)
```

```python
edges = cv2.Canny(image=img,threshold1=45,threshold2=80)
display(edges)
```
