```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%matplotlib inline
img = cv2.imread('Lane_01.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
shape = img.shape
shape = shape[:2]
shape
```

```python
width,height = 250,350

pts1= np.float32([[590,310],[692,300],[10,698],[1060,693]])
pts2 = np.float32([[0, 0], [width, 0], [0, height], [width, height]])

matrix = cv2.getPerspectiveTransform(pts1,pts2)

result = cv2.warpPerspective(img, matrix, (width,height))

plt.imshow(result)
                  
                  
```

```python
channel = result[:,:,1]
plt.imshow(channel)
```

```python
val , thresh = cv2.threshold(channel,130,255,cv2.THRESH_BINARY)
plt.imshow(thresh)
```
